```python
selected_df = df[['Accession', 'Release_Date','Species','Length','Host','Geo_Location','Collection_Date']]
#Rename the column
selected_df = selected_df.rename(columns={'Release_Date':'Release Date', 'Geo_Location':'Geo Location', 'Collection_Date':'Collection Date'})
```

```python
selected_df
```

```python
#Check the format of release date
selected_df['Collection Date'][10:50]
```

```python
#Keep only the date and remove the rest
selected_df['Collection Date'] = selected_df['Collection Date'].str[:10]
selected_df['Release Date'] = selected_df['Release Date'].str[:10]
```

```python
selected_df
```

```python
from bioinfokit.analys import Fasta
fasta_iter = Fasta.fasta_reader(file='nucleotide_jul20_jul22.fasta')
# read fasta file
data = []
for record in fasta_iter:
    header, sequence = record
    data.append([header[:-2], sequence])
print(data[0])
```

```python
fasta_df = pd.DataFrame(data, columns=['Accession', 'Nucleotide Fasta'])
selected_df = pd.merge(selected_df, fasta_df, on = ['Accession'], how='inner')
selected_df
```

```python
#Only keeps the rows after 
record_date = '2020-07-20'
#Select the genes after the given date
selected_df = selected_df[(selected_df['Collection Date'] > record_date)]
```

```python
selected_df = selected_df.reset_index()
del selected_df['index']
selected_df
```

```python
#Modify the string of Accession 
selected_df['Accession'] = selected_df['Accession'].apply(lambda x: "{}{}{}{}{}".format('<a href=\"https://www.ncbi.nlm.nih.gov/nuccore/',x,'?report=fasta\">',x,'</a>'))
```
